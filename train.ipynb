# Libraies and Data Sheet Lord #

```python
import pandas as pd 
import pickle
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor

### Data Set Lord ###
df=pd.read_csv('laptop_data.csv')
### Check data ##
df.head(5)
```

# Data Cleaning #

```python
## Ram line "Gb" Remove conver Integer ##
df['Ram']= df['Ram'].str.replace('GB','').astype(int)
df.head(5)  ##cheking##
```

# Train /Test spilit Separate  Features (X) and Target (y) #

```python
X = df[['Company','CPU','Ram','Storage','OpSys','Weight']]
y = df['Price']

## Saparate two  Data  Parts ##
X_train,X_test, y_train,y_test = train_test_split(X,y,test_size=0.15,random_state=42)
print(f"Training data row: {len(X_train)}, Testing data row:{(X_test)}")
```

# Create Ml Pipeline and Model Train #

```python
# Categorical  identify and transform #
categorical_features = ['Company','CPU','Ram','Storage','OpSys']
preprocessor= ColumnTransformer(
    transformers=[
        ('col_tnf',OneHotEncoder(sparse_output=False, drop='first',handle_unknown='ignore'),categorical_features) 
    ],
    remainder='passthrough'
)

# Create Random Forest Model With Pipeline #
model=RandomForestRegressor(n_estimators=100,random_state=42)
pipe=Pipeline([
    ('step1',preprocessor),
    ('step2',model)
])

# Train model #
pipe.fit(X_train,y_train)

print("✅ Model Trained Successfully!")
print(f"📈 Training Accuracy (R2 Score): {pipe.score(X_train, y_train):.2f}")
```

# Model Save pickel file #

```python
with open('laptop_pipe.pkl','wb')as f:
    pickle.dump(pipe,f)
print("💾 Model saved as 'laptop_pipe.pkl'")
```
